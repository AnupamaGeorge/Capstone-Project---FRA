```python
plt.figure(figsize = (18,8))
sns.countplot(df['age'])
plt.axhline(y=150, color = 'r')
plt.show()
```

Here, we can say that calls have been made to the customers of age from 18 to 95 years. \
However, most of the call have been made to customers in the age group of 23 years to 60 years. \
So, it can be conluded that bank tried to target working class customers.

```python
success_percent = df[df['y']=='yes']['age'].value_counts()*100/df['age'].value_counts()
plt.figure(figsize = (18,8))
plt.plot(success_percent)
plt.axhline(y=10, color = 'r')
plt.axvline(x=23, color = 'r')
plt.axvline(x=60, color = 'r')
plt.xlabel('age')
plt.ylabel('percentage of subscribing term deposit w.r.t. number of call made')
plt.show()
```
