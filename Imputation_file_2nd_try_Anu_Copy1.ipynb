```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("bank-full.csv", sep = ';')
df.head()
```

```python
df1= df.copy(deep=True)
df1['poutcome'].value_counts()
```

```python
df1.loc[(df1.poutcome=='unknown') & (df1.pdays==-1),'poutcome'] = 'Not Contacted'
```

```python
df1['poutcome'].value_counts()
```

```python
df1.replace('unknown',np.NaN, inplace=True)
df1
```

```python
df1.isnull().sum()
```

```python
X = df1.drop(['y'], axis = 1)
X
```

------------------------

# ONEHOT ENCODING BLOCK NOT CONSIDERING NOW

```python
pd.set_option('display.max_columns', None)
X=pd.get_dummies(X, columns=['job','marital','default','housing','loan','contact','poutcome'], dummy_na = True, drop_first=True)
X
```

```python
X.drop(['marital_nan', 'default_nan','housing_nan', 'loan_nan'], axis=1, inplace=True)
```

```python
X['job_nan'].replace(to_replace = 1, value = np.NaN, inplace =True)
X['contact_nan'].replace(to_replace = 1, value = np.NaN, inplace =True)
X['poutcome_nan'].replace(to_replace = 1, value = np.NaN, inplace =True)
```

```python
X
```

----------------------------------------------

# ORDINAL ENCODING BLOCK

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
columns=['job', 'education','marital','default','month','housing','loan','contact','poutcome']
# Create an empty dictionary ordinal_enc_dict
ordinal_enc_dict = {}

for col_name in columns:
    # Create Ordinal encoder for col
    ordinal_enc_dict[col_name] = OrdinalEncoder()
    col = X[col_name]
    
    # Select non-null values of col
    col_not_null = col[col.notnull()]
    reshaped_vals = col_not_null.values.reshape(-1, 1)
    encoded_vals = ordinal_enc_dict[col_name].fit_transform(reshaped_vals)
    
    # Store the values to non-null values of the column in users
    X.loc[col.notnull(), col_name] = np.squeeze(encoded_vals)
```

```python
X
```

```python
X.drop('col_name',axis=1,inplace=True)
```

```python
X
```

# KNN Imputation

```python
from sklearn.impute import KNNImputer

term_knn_imputed = X.copy()
knn_imputer = KNNImputer(n_neighbors=5, weights="uniform")
```

```python
term_knn_imputed.iloc[:, :] = np.round(knn_imputer.fit_transform(term_knn_imputed))
```

```python
pd.set_option('display.max_columns', None)
term_knn_imputed.head()
```

```python
term_knn_imputed['education'].value_counts()
```

# MICE Imputation

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
term_mice_imputed = X.copy()
mice_imputer = IterativeImputer(initial_strategy = 'most_frequent', imputation_order = 'random')
```

```python
term_mice_imputed.iloc[:, :] = np.round(mice_imputer.fit_transform(term_mice_imputed))
```

```python
cols=['job','education','contact', 'poutcome']
for i in cols:
  print(term_mice_imputed[i].value_counts())
```
